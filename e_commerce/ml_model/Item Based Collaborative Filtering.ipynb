```python
user_product_df = df.pivot_table(index=["User_id"], columns=["Product Name"], values="Rating").fillna(0)

user_product_df.shape
```

```python
user_product_df.head(10)
```

```python
product_name = "Baluchari Saree 2"
product_name = user_product_df[product_name]
```

```python
suggestions=user_product_df.corrwith(product_name).sort_values(ascending=False)
```

```python
X = user_product_df.values.T
X.shape
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=10, random_state=42)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```

```python
product_title = user_product_df.columns
product_title_list = list(product_title)
# print(product_title_list)
coffey_hands = product_title_list.index("Teracotta 1")
```

```python
def recom_ItemBase():
    corr_coffey_hands  = corr[coffey_hands]
    return list(product_title[(corr_coffey_hands >= 0.7)])
```

```python
import pickle
# Saving model to disk
pickle.dump(recom_ItemBase(), open('ItemBase.pkl','wb'))
```
